```
#@title Default title text
#Pandas is a Data Analysis library that creates DataFrames
#Most external libraries need to be downloaded using pip
import pandas as pd
#my_instance=MyClass() traditional syntax of instantiating a class
df=pd.read_csv("2019.csv")
df.head()#first rows of DataFrame

df.drop(["Rk","2PM","2PP","FantPt","DKPt","FDPt","VBD","PosRank","OvRank","PPR","Fmb","GS"], axis=1, inplace=True)
#axis 1 is horizontal columns and axis=0 is vertical rows
#inplace=True saves our changes, without it, our drops would still be shown

df["Player"]=df["Player"].apply(lambda x: x.split("*")[0]).apply(lambda x: x.split("\\")[0])
#When clicking on excel spreadsheet, most players name had a special character that was hidden such a Ezeikiel Elliot*\EllEzz00.
#df["Column Name"]
#df["column"]=df["column"].apply(transforming_function)
#lambda sort of creates its own argument to be used
#.apply allows program to run code on every line 
df.rename({
    "TD":"PassingTD",
    "TD.1":"RushingTD",
    "TD.2":"ReceivingTD",
    "TD.3":"TotalTD",
    "Yds":"PassingYDs",
    "Yds.1":"RushingYDs",
    "Yds.2":"ReceivingYDs",
    "Att":"PassingAtt",
    "Att.1":"RushingAtt"

}, axis=1, inplace=True)

df["FantasyPoints"]=df["PassingYDs"]/25 +df["PassingTD"]*4 - df["Int"]*2 + df["Rec"]+df["ReceivingYDs"]/10+df["ReceivingTD"]*6 + df["RushingYDs"]/10 + df["RushingTD"]*6

draft=100
#be careful with : and =
#new variable to take out first 100 draft numbers
#doesn't matter if draft_pool_limit(100), as long as its referenced later in code

def calculate_average_based_off_position(our_df,pos):
  return our_df[our_df["FantPos"]== pos]["FantasyPoints"].mean()
  #FantPos is to be used in the final product 
  #(our_df,pos) and ==pos are related for the confused me
  #our_df[our_df["some_column"]=="some value"]
  #We are telling python to give me our DataFrame where the value FantPos equals the pos given
  #== is a boolean expression
  #.mean() calculates the mean of the FantasyPoints column in excel 

cabop=calculate_average_based_off_position
#renamed to cabop

new_df=df.sort_values(by=["FantasyPoints"],ascending=False)[:draft]
#[:draft] could also be used .head(),[:100], or [0,100]
#new_df is a new variable unused and is the final code used for output

rb_avg_replacement=cabop(new_df, "RB")
wr_avg_replacement=cabop(new_df, "WR")
te_avg_replacement=cabop(new_df, "TE")
#avg_replacement are variables

pos_replacement_vals={
    "RB": rb_avg_replacement,
    "WR": wr_avg_replacement,
    "TE": te_avg_replacement
    }

new_df= pd.DataFrame({}, columns=["Player", "FantPos","FantasyPoints","VOR"])
#new_df does not include the earlier new_df, in order to reuse old new_df, you'd have to rename it
#new variable, new DataFrame created
#this new_df creates the columns

for position, replacement_value in pos_replacement_vals.items():
    pos_df=df[df["FantPos"]==position]
    pos_df=pos_df[["Player","FantPos","FantasyPoints"]]
    pos_df["VOR"]=pos_df["FantasyPoints"]  - replacement_value
    new_df=pd.concat([new_df, pos_df])
#Using the dictionary made earlier
#pos_df is filtering out the positions
#.concat adds new_df and pos_df
#replacement_value I think is a given name for pos_replacement_value
new_df.sort_values(by=["VOR"],ascending=False).head(25)



```
